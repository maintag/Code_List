```python
class QLearning:
    def __init__(self, alpha=0.5, gamma=1):
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = 0.99
        self.q_table = {}
        self.rewards_table = {}
        
    def get_q_value(self, state, action):
        state_str = str(state).replace(" ", "")
        action_str = str(action).replace(" ", "")
        state_action = (state_str, action_str)
        if state_action not in self.q_table:
            self.q_table[state_action] = 0
            self.rewards_table[state_action] = 0
        return self.q_table[state_action]
    
    def choose_action(self, state):
        if np.random.uniform() < self.epsilon:
            self.epsilon *=0.9
            return Action()
        else:
            actions = [Action() for _ in range(100)]
            q_values = [self.get_q_value(state, action) for action in actions]
            max_q_value = max(q_values)
            if q_values.count(max_q_value) > 1:
                max_q_value = max(self.get_q_value(state, a) for a in actions)
                best_action = [a for a in actions if self.get_q_value(state, a) == max_q_value][0]
                return best_action
            else:
                return actions[np.argmax(q_values)]

    def learn(self, state, action, reward, next_state):
        state_str = str(state).replace(" ", "")
        action_str = str(action).replace(" ", "")
        state_action = (state_str, action_str)

        td_target = reward + self.gamma * max([self.get_q_value(next_state, Action()) for _ in range(10)])
        td_error = td_target - self.get_q_value(state, action)
        self.q_table[state_action] += self.alpha * td_error
        self.rewards_table[state_action] = reward
        
    def get_reward(self, state_ks, new_state_ks):
        if new_state_ks < state_ks:
            return 1-(new_state_ks)
        else:
            return -1
    
    def get_q_table_as_dataframe(self):
        q_table_dict = {'State': [], 'Action': [], 'Q_Value': [], 'K/S':[]}
        for state_action, q_value in self.q_table.items():
            state, action = state_action
            q_table_dict['State'].append(state)
            q_table_dict['Action'].append(action)
            q_table_dict['Q_Value'].append(q_value)
            q_table_dict['K/S'].append(1- self.rewards_table[state_action])
        return pd.DataFrame(q_table_dict)
        
q_learning  = QLearning()

```

```python
class QValue:
    def __init__(self, n_episodes, n_iterations, state_action_attrs):
        self.n_episodes = n_episodes
        self.n_iterations = n_iterations
        self.state_action_attrs = state_action_attrs
        self.best_state = None
        self.best_ks_value = float('inf')
        self.best_e_value = float('inf')
        self.best_iteration = None
        self.episode_data = []
        
    def run(self, q_learning):
        for episode in range(self.n_episodes):
            state = State()
            action = q_learning.choose_action(state)
            new_state = action.apply(state)
            reward = q_learning.get_reward(state.get_ks(), new_state.get_ks())
            q_learning.learn(state, action, reward, new_state)
            state = new_state
            
            episode_data = {'Episode': episode+1, 
                'New_KS_state': str(new_state), 
                'epsilon': q_learning.epsilon,
                'KS_Value': new_state.get_ks(),
                'Reward': reward}
            
            self.episode_data.append(episode_data)


            
        return pd.DataFrame(self.episode_data)
```

```python
q_learning = QLearning(alpha=0.5, gamma=0.9)
q_value = QValue(n_episodes= 1000, n_iterations=10, state_action_attrs=['상승속도1', '상승속도2', '상승속도3', '상승온도3', '상승온도유지시간3'])
df = q_value.run(q_learning)
print("Best state:", q_value.best_state)
print("Best KS:", q_value.best_ks_value)
print("Best iteration:", q_value.best_iteration)
df=df[df.columns[1:]]
#df.to_csv("C:/Users/leewa/Desktop/잔류01%.csv", encoding='euc-kr')
```
