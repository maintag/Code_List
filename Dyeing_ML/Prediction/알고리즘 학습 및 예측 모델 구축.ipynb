# Train - Test 차이

```python
RFregressor = RandomForestRegressor(max_depth=20, random_state= 0)
RFregressor.fit(x_train,y_train)
y_pred = RFregressor.predict(x)    
print("training set score: {:.4f}".format(RFregressor.score(x_train, y_train))) # 훈련셋 결정계수
print("test set score: {:.4f}".format(RFregressor.score(x_test, y_test))) # 테스트셋 결정계수
print("r^2 score: {:.4f}".format(r2_score(y, y_pred))) # r2스코어
print("MAE score: {:.4f}".format(mean_absolute_error(y, y_pred))) # MAE스코어
print("MSE score: {:.4f}".format(mean_squared_error(y, y_pred))) # MSE스코어

# 훈련 세트에서의 예측
y_train_pred = RFregressor.predict(x_train)

# 테스트 세트에서의 예측
y_test_pred = RFregressor.predict(x_test)

# 훈련 세트의 성능 평가
train_r2 = r2_score(y_train, y_train_pred)
train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)

# 테스트 세트의 성능 평가
test_r2 = r2_score(y_test, y_test_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

# 결과 출력
print("Training set - R2 Score: {:.4f}, MAE: {:.4f}, MSE: {:.4f}".format(train_r2, train_mae, train_mse))
print("Test set - R2 Score: {:.4f}, MAE: {:.4f}, MSE: {:.4f}".format(test_r2, test_mae, test_mse))
```

# K-Fold

```python
GB = GradientBoostingRegressor(n_estimators = 500,max_depth=8, min_samples_split=7,random_state=0)
GB.fit(x_train,y_train)
y_pred = GB.predict(x)    
print("training set score: {:.2f}".format(GB.score(x_train, y_train))) # 훈련셋 결정계수
print("test set score: {:.2f}".format(GB.score(x_test, y_test))) # 테스트셋 결정계수
print("r^2 score: {:.2f}".format(r2_score(y, y_pred))) # r2스코어
print("MAE score: {:.2f}".format(mean_absolute_error(y, y_pred))) # MAE스코어
print("MSE score: {:.2f}".format(mean_squared_error(y, y_pred))) # MSE스코어

k = 5  # 원하는 k-fold의 값으로 변경
kf = KFold(n_splits=k, shuffle=True, random_state=0)

# 교차 검증을 통한 R^2 스코어 계산
r2_scores = cross_val_score(GB, x, y, cv=kf, scoring='r2')

# 교차 검증을 통한 MAE 스코어 계산
mae_scores = -cross_val_score(GB, x, y, cv=kf, scoring='neg_mean_absolute_error')

# 교차 검증을 통한 MSE 스코어 계산
mse_scores = -cross_val_score(GB, x, y, cv=kf, scoring='neg_mean_squared_error')

# R^2 스코어의 평균과 표준 편차 계산
mean_r2 = r2_scores.mean()
std_r2 = r2_scores.std()

# MAE 스코어의 평균과 표준 편차 계산
mean_mae = mae_scores.mean()
std_mae = mae_scores.std()

# MSE 스코어의 평균과 표준 편차 계산
mean_score = mse_scores.mean()
std_score = mse_scores.std()

print("K-Fold R^2 score: {:.2f}".format(mean_r2))
print("Standard deviation of R^2: {:.2f}".format(std_r2))
print("K-Fold MAE score: {:.2f}".format(mean_mae))
print("Standard deviation of MAE: {:.2f}".format(std_mae))
print("K-Fold MSE score: {:.2f}".format(mean_score))
print("Standard deviation: {:.2f}".format(std_score))
```
