```python
def grid_data(input_value):
    ##공정 데이터 부문 그리드서치 데이터 생성##
    #'Lab 염색 시작온도',
    num_cor_slice1 = [40]
    #'Lab 염색 상승속도 #1',
    num_cor_slice2 = [0.5,1.5]
    #'Lab 염색 상승온도 #1', 
    num_cor_slice3 = [64]
    #'Lab 염색 상승온도 #1 유지시간', 
    num_cor_slice4 = [0]
    #'Lab 염색 상승속도 #2',
    num_cor_slice5 = [0.5,0.75,1.5]
    #'Lab 염색 상승온도 #2',
    num_cor_slice6 = [100]
    #'Lab 염색 상승온도 #2 유지시간', 
    num_cor_slice7 = [0]
    #'Lab 염색 상승속도 #3',
    num_cor_slice8 = [0.5,0.833,1.25,1.458,1.5,1.667]
    #'Lab 염색 상승온도 #3',
    num_cor_slice9 = [120,130,135,140]
    #'Lab 염색 상승온도 #3 유지시간', 
    num_cor_slice10 = [10,20,30,40,50,60,70,80]
    #'Lab 염색 하강속도 #1',
    num_cor_slice11 = [1.97222]
    #'Lab 염색 하강온도 #1',
    num_cor_slice12 = [64]
    #'Lab 염색 하강온도 #1 유지시간', 
    num_cor_slice13 = [0]
    #'Lab 염색 하강속도 #2',
    num_cor_slice14 = [0]
    #'Lab 염색 하강온도 #2',
    num_cor_slice15 = [64]
    #'Lab 염색 하강온도 #2 유지시간', 
    num_cor_slice16 = [0]
    #'Lab 염색 하강속도 #3',
    num_cor_slice17 = [0]
    #'Lab 염색 하강온도 #3', 
    num_cor_slice18 = [64]
    #'Lab 염색 하강온도 #3 유지시간',
    num_cor_slice19= [0]
    #'Lab 염색 종료속도',
    num_cor_slice20 = [0]
    #'Lab 염색 종료온도', 
    num_cor_slice21 = [64]
    #'Lab 염색 종료온도 유지시간'
    num_cor_slice22= [0]
    #경우의수 곱 구하기
    num_linspace = []
    num_linspace.append(num_cor_slice1)
    num_linspace.append(num_cor_slice2)
    num_linspace.append(num_cor_slice3)
    num_linspace.append(num_cor_slice4)
    num_linspace.append(num_cor_slice5)
    num_linspace.append(num_cor_slice6)
    num_linspace.append(num_cor_slice7)
    num_linspace.append(num_cor_slice8)
    num_linspace.append(num_cor_slice9)
    num_linspace.append(num_cor_slice10)
    num_linspace.append(num_cor_slice11)
    num_linspace.append(num_cor_slice12)
    num_linspace.append(num_cor_slice13)
    num_linspace.append(num_cor_slice14)
    num_linspace.append(num_cor_slice15)
    num_linspace.append(num_cor_slice16)
    num_linspace.append(num_cor_slice17)
    num_linspace.append(num_cor_slice18)
    num_linspace.append(num_cor_slice19)
    num_linspace.append(num_cor_slice20)
    num_linspace.append(num_cor_slice21)
    num_linspace.append(num_cor_slice22)
    
    #경우의수 데이터 생성
    cartesian_cor = list(itertools.product(*num_linspace))
    
    #데이터프레임 형태변환
    ca=pd.DataFrame(cartesian_cor)
    ca.columns=['Lab 염색 시작온도', 'Lab 염색 상승속도 #1','Lab 염색 상승온도 #1', 'Lab 염색 상승온도 #1 유지시간', 
                'Lab 염색 상승속도 #2','Lab 염색 상승온도 #2', 'Lab 염색 상승온도 #2 유지시간', 
                'Lab 염색 상승속도 #3','Lab 염색 상승온도 #3', 'Lab 염색 상승온도 #3 유지시간', 
                'Lab 염색 하강속도 #1','Lab 염색 하강온도 #1', 'Lab 염색 하강온도 #1 유지시간', 
                'Lab 염색 하강속도 #2','Lab 염색 하강온도 #2', 'Lab 염색 하강온도 #2 유지시간', 
                'Lab 염색 하강속도 #3','Lab 염색 하강온도 #3', 'Lab 염색 하강온도 #3 유지시간',
                'Lab 염색 종료속도','Lab 염색 종료온도', 'Lab 염색 종료온도 유지시간']
    
    
    #입력데이터 전처리
    ex_data=pd.DataFrame(input_value,index=[0]) #데이터프레임화
    
    #염료비율 변수 추가
    color_mat = ex_data[['Dorosperse Red KKL', 'Dorosperse Blue KKL', 'Dorosperse B/K KKL',
       'Dorosperse Dark Grey KKL', 'Dorosperse Brown K-3LR',
       'Dianix Yellow AM-2R', 'Dianix Red AM-SLR', 'Dianix Blue AM-2G',
       'Dianix Black AM-SLR', 'Dianix Grey AM-SLR', 'Dianix Yellow Brown AM-R',
       'Dorosperse Yellow KKL', 'Dorosperse Black KKL']].values
    
    col = ['Dorosperse Red KKL', 'Dorosperse Blue KKL', 'Dorosperse B/K KKL',
       'Dorosperse Dark Grey KKL', 'Dorosperse Brown K-3LR',
       'Dianix Yellow AM-2R', 'Dianix Red AM-SLR', 'Dianix Blue AM-2G',
       'Dianix Black AM-SLR', 'Dianix Grey AM-SLR', 'Dianix Yellow Brown AM-R',
       'Dorosperse Yellow KKL', 'Dorosperse Black KKL']
    
    ex_data[[i+"_prop" for i in col]] =color_mat/color_mat.sum(axis=1).reshape(-1,1)
    
    #조제변수들 위치 끝으로 이동
    ex_data=ex_data[['Dorosperse Red KKL', 'Dorosperse Blue KKL', 'Dorosperse B/K KKL',
       'Dorosperse Dark Grey KKL', 'Dorosperse Brown K-3LR',
       'Dianix Yellow AM-2R', 'Dianix Red AM-SLR', 'Dianix Blue AM-2G',
       'Dianix Black AM-SLR', 'Dianix Grey AM-SLR', 'Dianix Yellow Brown AM-R',
       'Dorosperse Yellow KKL', 'Dorosperse Black KKL','Dorosperse Red KKL_prop', 'Dorosperse Blue KKL_prop', 'Dorosperse B/K KKL_prop',
       'Dorosperse Dark Grey KKL_prop', 'Dorosperse Brown K-3LR_prop',
       'Dianix Yellow AM-2R_prop', 'Dianix Red AM-SLR_prop', 'Dianix Blue AM-2G_prop',
       'Dianix Black AM-SLR_prop', 'Dianix Grey AM-SLR_prop', 'Dianix Yellow Brown AM-R_prop',
       'Dorosperse Yellow KKL_prop', 'Dorosperse Black KKL_prop','배합_Sunsolt RM-340S', '배합_빙초산']]
    
    #입력변수 완성
    ex_data2=pd.DataFrame(np.repeat(np.array(ex_data),len(ca),axis=0))
    ex_data2.columns=ex_data.columns
    
    ##입력데이터와 공정데이터 결합##
    sim_data=pd.concat([ex_data2,ca],axis=1)
    
    ##train데이터 생성##
    x_np=np.array(sim_data)

        #가상 변수 생성
    zero=0
    x_simul=np.repeat(zero,[x_np.shape[0]])
    x_simul=np.array([x_simul])
    x_simul=x_simul.reshape(x_simul.shape[1],x_simul.shape[0])

        #X와 X_simul 결합
    x_train=np.concatenate([x_np,x_simul],axis=1)
    #잔욕농도예측모델 사용 X변수중 train데이터 생성
    x_train_ks=sim_data
   
    #예측 데이터 생성
    Y_pred=pd.DataFrame(model_KS.predict(x_train_ks))
    
    sim_data=pd.concat([sim_data,Y_pred],axis=1)
    
    return sim_data
```
