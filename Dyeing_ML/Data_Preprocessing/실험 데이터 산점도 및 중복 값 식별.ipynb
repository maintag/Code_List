# 동일 배합에서 공정 설정 별 잔류염료 매핑

```python
a = {}

for idx2, row2 in data2.iterrows():
    val = data2.iloc[idx2]
    a[f'Formulation_{idx2}'] = []  # 변수명을 동적으로 생성

    for idx, row in data.iterrows():
        condition = [
            (data.iloc[idx]['Dorosperse Red KKL'] == val['Dorosperse Red KKL']),
            (data.iloc[idx]['Dorosperse Blue KKL'] == val['Dorosperse Blue KKL']),
            (data.iloc[idx]['Dorosperse B/K KKL'] == val['Dorosperse B/K KKL']),
            (data.iloc[idx]['Dorosperse Dark Grey KKL'] == val['Dorosperse Dark Grey KKL']),
            (data.iloc[idx]['Dorosperse Brown K-3LR'] == val['Dorosperse Brown K-3LR']),
            (data.iloc[idx]['Dianix Yellow AM-2R'] == val['Dianix Yellow AM-2R']),
            (data.iloc[idx]['Dianix Red AM-SLR'] == val['Dianix Red AM-SLR']),
            (data.iloc[idx]['Dianix Blue AM-2G'] == val['Dianix Blue AM-2G']),
            (data.iloc[idx]['Dianix Black AM-SLR'] == val['Dianix Black AM-SLR']),
            (data.iloc[idx]['Dianix Grey AM-SLR'] == val['Dianix Grey AM-SLR']),
            (data.iloc[idx]['Dianix Yellow Brown AM-R'] == val['Dianix Yellow Brown AM-R']),
            (data.iloc[idx]['Dorosperse Yellow KKL'] == val['Dorosperse Yellow KKL']),
            (data.iloc[idx]['Dorosperse Black KKL'] == val['Dorosperse Black KKL'])
        ]
        if all(condition):
            a[f'Formulation_{idx2}'].append(data.iloc[idx]['잔욕염색 검사_K/S'])
```

# 배합 별 잔류염료 값 테이블

```python
df = pd.DataFrame.from_dict(a, orient='index')

directory_path = 'C:/Users/leewa/Desktop/'
os.makedirs(directory_path, exist_ok=True)
#df.to_csv(os.path.join(directory_path, '배합 당 KS 정보.csv'), encoding='euc-kr')
```
